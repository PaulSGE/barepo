```python
### Access dataframe
### -----------------------------------------------------

mask_df = pd.read_csv("/content/drive/MyDrive/ColabNotebooks/Tesla_Model_3/padded.csv")
```

```python
### Create hit count for battery capacity when = 1
### -----------------------------------------------------

def hit(batch , coloum):
  hit_count = batch.index[0] - 1
  for i in batch[coloum]:
    hit_count += 1
    if i == 1:
      return hit_count
  raise Exception("No entries Found")
  return -1
```

```python
### Create output list
### -----------------------------------------------------

def mask(batch, batch_no , coloum):
  output = []
  for i in range(hit(batch,coloum) +1  , len(batch['count']) + batch.index[0] , 1):
    output.append(i)
  return output
```

```python
### Make copy of mask_df

test_df = mask_df.copy()
```

```python
### 30% battery capacity masking script
### -----------------------------------------------------

grp = test_df.groupby('batch_id')
for i in test_df['batch_id'].unique():
  try:
    batch_df = grp.get_group(i)
    mask_output = mask(batch_df,i-1,'30_battery')
    for id in mask_output:
      test_df.at[id, '20_battery'] =  0
      test_df.at[id, '30_battery'] = 0
      test_df.at[id, '40_battery'] = 0
      test_df.at[id, 'range_km'] =  0
      test_df.at[id, 'battery_Wh'] = 0
      test_df.at[id, 'Wh_Km'] = 0
      test_df.at[id, 'trip_km'] = 0
      test_df.at[id, 'pickup'] = 0
      test_df.at[id, 'dropoff'] = 0
      test_df.at[id, 'count'] = 0
      test_df.at[id, 'batch_id'] = 0
      test_df.at[id, 'b1'] =  0
      test_df.at[id, 'b2'] = 0
      test_df.at[id, 'b3'] = 0
      test_df.at[id, 'b4'] = 0
      test_df.at[id, 'start_daymonth'] =  0
      test_df.at[id, 'start_dayday'] = 0
      test_df.at[id, 'start_dayhour'] = 0
      test_df.at[id, 'end_daymonth'] = 0
      test_df.at[id, 'end_dayday'] = 0
      test_df.at[id, 'end_dayhour'] = 0
  except:
    print(f"No entries found in the batch {i}")


test_df.to_csv('/content/drive/MyDrive/ColabNotebooks/Tesla_Model_3/mask30.csv', index=False)
```
